```python
from PIL import Image, ImageDraw

# --- 1. Definisikan Nama File ---
# Ganti nama file di sini jika berbeda di lingkungan Anda
BACKGROUND_IMAGE = "batik pekalongan.jpeg"
FOREGROUND_IMAGE = "kura-kura.jpeg"
OUTPUT_IMAGE = "natural_style_montase.jpeg"

# --- 2. Fungsi Utama untuk Memproses Gambar ---
def create_natural_montage(bg_path, fg_path, output_path):
    try:
        # Muat kedua gambar
        background = Image.open(bg_path).convert("RGB")
        foreground = Image.open(fg_path).convert("RGBA") # Convert ke RGBA untuk transparansi

        # --- 3. Atur Ukuran dan Posisi ---

        # Tentukan ukuran akhir (misalnya, ukuran latar belakang)
        bg_width, bg_height = background.size

        # Hitung ukuran baru untuk kura-kura (misalnya 70% dari tinggi latar belakang)
        fg_ratio = 0.70
        new_fg_height = int(bg_height * fg_ratio)
        new_fg_width = int(foreground.width * (new_fg_height / foreground.height))

        # Ubah ukuran kura-kura
        foreground = foreground.resize((new_fg_width, new_fg_height))

        # Tentukan posisi di tengah bawah
        position_x = (bg_width - new_fg_width) // 2
        position_y = bg_height - new_fg_height - 50 # 50 piksel dari bawah

        # --- 4. Penyesuaian Gaya Natural (Transparansi/Opasitas) ---

        # 4a. Membuat lapisan latar depan yang agak transparan
        # Kami akan membuat lapisan kura-kura sedikit transparan (misalnya, 80% opasitas)
        alpha = foreground.getchannel('A')
        alpha = alpha.point(lambda p: p * 0.8) # Mengurangi opasitas menjadi 80%
        foreground.putalpha(alpha)

        # 4b. Opsional: Penyesuaian warna (membuat kura-kura lebih menyatu)
        # Kami akan melakukan sedikit desaturasi pada kura-kura
        foreground_rgb = foreground.convert("RGB")
        from PIL import ImageEnhance
        enhancer = ImageEnhance.Color(foreground_rgb)
        foreground_desaturated = enhancer.enhance(0.9) # Mengurangi saturasi sebesar 10%

        # Gabungkan kembali dengan alpha channel yang sudah dimodifikasi
        foreground_final = foreground_desaturated.convert("RGBA")
        foreground_final.putalpha(alpha)

        # --- 5. Montase/Gabungkan Gambar ---

        # Tempelkan gambar kura-kura yang sudah diubah ukurannya dan transparan
        # ke gambar batik di posisi yang telah ditentukan.
        background.paste(foreground_final, (position_x, position_y), foreground_final)

        # --- 6. Simpan Hasil ---
        background.save(output_path)
        print(f"✅ Gambar gaya natural berhasil dibuat dan disimpan sebagai: {output_path}")

    except FileNotFoundError:
        print("❌ ERROR: Pastikan kedua file gambar (kura-kura.jpeg dan batik pekalongan.jpeg) ada di direktori yang sama dengan skrip Python.")
    except Exception as e:
        print(f"❌ Terjadi kesalahan: {e}")

# --- Panggil Fungsi ---
if __name__ == "__main__":
    create_natural_montage(BACKGROUND_IMAGE, FOREGROUND_IMAGE, OUTPUT_IMAGE)
```
